```python
# TODO: Add import statements
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Assign the data to predictor and outcome variables
# TODO: Load the data
train_data = pd.read_csv('data.csv', header = None)
X = train_data.iloc[:,:-1]
y = train_data.iloc[:,-1]
```

```python
# TODO: Create the linear regression model with lasso regularization.
lasso_reg = Lasso(alpha=1)

# TODO: Fit the model.
lasso_reg.fit(X, y)

# TODO: Retrieve and print out the coefficients from the regression model.
reg_coef = lasso_reg.coef_
reg_coef
```

```python
alphas = [0, 0.1 , 1 , 10 , 100 , 1000]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
for i in alphas:
    
    lasso_reg = Lasso(alpha=i)
    lasso_reg.fit(X_scaled, y)
    Lasso_reg_coef = lasso_reg.coef_ 

    ridge_reg = Ridge(alpha= i)
    ridge_reg.fit(X_scaled , y)
    ridge_reg_coef = ridge_reg.coef_ 
    
    fig, axes = plt.subplots(1, 2)
    axes[0].bar(range(len(Lasso_reg_coef)), Lasso_reg_coef , color = 'red')
    axes[0].set_title('Lasso alpha equal to: ' + str(i))
    
    
    axes[1].bar(range(len(ridge_reg_coef)), ridge_reg_coef , color= 'green')
    axes[1].set_title('Ridge alpha equal to: ' + str(i))
    
    plt.show()
```
